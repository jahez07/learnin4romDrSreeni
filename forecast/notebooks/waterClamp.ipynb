# **Testing Prophet on Water Clamp Data**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

plt.style.use('ggplot')
plt.style.use('fivethirtyeight')
```

```python
clmp = pd.read_csv(
    "../data/waterClamp3d.csv",
)
```

```python
clmp = clmp.rename(columns={'timestamp':'Datetime'})
clmp1 = clmp.set_index('Datetime')
```

```python
clmp1.index
```

```python
clmp1.index = pd.to_datetime(clmp1.index, format='mixed', utc=True, infer_datetime_format=True)
```

```python
clmp1.index
```

```python
import seaborn as sns
color_pal = sns.color_palette()

clmp1.plot(
    style='.',
    figsize=(10,5), 
    ms=1, 
    color=color_pal[0],
    title="Clamp Water"
)

plt.show()
```
